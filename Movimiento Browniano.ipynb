```python
stock = pd.read_csv('./data/SPY.csv')
```

```python
plt.plot(stock['Close'])
```

```python
stock
```

```python
diferencias = stock['Close'] - stock['Close'].shift()
```

```python
plt.plot(diferencias)
```

```python
media_muestral = diferencias.mean()
media_muestral
```

```python
desv_muestral = diferencias.std()
desv_muestral
```

```python
mod1 = np.random.normal(media_muestral,desv_muestral,252)
mod2 = np.random.normal(media_muestral,desv_muestral,252)
mod3 = np.random.normal(media_muestral,desv_muestral,252)
mod4 = np.random.normal(media_muestral,desv_muestral,252)
mod5 = np.random.normal(media_muestral,desv_muestral,252)
```

```python
modelaciones = []
n = 100
for i in range(n):
    m = np.random.normal(media_muestral,desv_muestral,252)
    modelaciones.append(m)
```

```python
plt.figure()

for m in modelaciones:
    plt.plot(m)

plt.show()
```

```python
modelacion_1 = stock['Close'][0] + mod1.cumsum()
modelacion_2 = stock['Close'][0] + mod2.cumsum()
modelacion_3 = stock['Close'][0] + mod3.cumsum()
modelacion_4 = stock['Close'][0] + mod4.cumsum()
modelacion_5 = stock['Close'][0] + mod5.cumsum()
```

```python
plt.figure(figsize = (12,8))
plt.title('VALOR REAL DE ACCION VS MODELACIONES')
plt.xlabel('Tiempo')
plt.ylabel('Precio')

plt.plot(stock['Close'],linewidth=3,label='Valor Real')
plt.plot(modelacion_1,label='Modelacion 1')
plt.plot(modelacion_2,label='Modelacion 2')
plt.plot(modelacion_3,label='Modelacion 3')
#plt.plot(modelacion_4,label='Modelacion 4')
#plt.plot(modelacion_5,label='Modelacion 5')

plt.legend()
plt.show()
```
